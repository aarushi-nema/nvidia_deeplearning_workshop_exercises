Here we will perform data preparation, model creation, and model training on a data set of American Sign Language images.


#Loading the data

We will be loading custom data. The sign language data is in CSV (Comma Separated Values) format which is the same data structure as google sheets and microsoft excel. 

We will use pandas to load and manipulate the data. 

```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
```

```python
import io
import matplotlib.pyplot as plt
```

```python
#We will use the read_csv method from the pandas api to load the csv file and store it as a "DataFrame"
training_dataframe = pd.read_csv(io.StringIO(uploaded['sign_mnist_train.csv'].decode('utf-8')))
validation_dataframe = pd.read_csv(io.StringIO(uploaded['sign_mnist_test.csv'].decode('utf-8')))
```

We can use the head() function to view the first few entries in the data frames.

```python
training_dataframe.head()
```

Now we will remove the labels coumn from the data frame and add it to the y_train and y_valid array. 

```python
y_train = training_dataframe['label']
y_valid = validation_dataframe['label']
del training_dataframe['label']
del validation_dataframe['label']
```
